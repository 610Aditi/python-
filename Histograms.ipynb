```python
blood_sugar_men = [113,83,90,160,98,135,77,83,129,69] 
blood_sugar_women = [120,80,75,130,95,156,85,70,105,55] 

plt.xlabel("sugar range")
plt.ylabel("Total no of patients")
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95 ,color =['orange','purple'],label=["men","women"],orientation="horizontal")
plt.legend()
```
