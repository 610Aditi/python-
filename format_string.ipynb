```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,58,48,49,46]
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y,'rD')
plt.show()
```
